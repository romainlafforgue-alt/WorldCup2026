```python
from understatapi import UnderstatClient
import pandas as pd

# Liste des ligues que vous voulez analyser
ligues = ["EPL", "Bundesliga", "Serie_A", "Ligue_1", "La_Liga"]

understat = UnderstatClient()
all_data = []

# Boucle pour récupérer chaque championnat
for nom_ligue in ligues:
    print(f"Récupération de {nom_ligue}...")
    data = understat.league(league=nom_ligue).get_player_data(season="2025")
    
    # Transformation en DataFrame
    df_temp = pd.DataFrame(data)
    df_temp['league'] = nom_ligue # Très important pour ne pas mélanger les joueurs !
    all_data.append(df_temp)

# Fusionner le tout en un seul grand DataFrame
df_final = pd.concat(all_data, ignore_index=True)

# Résultat final
display(df_final.sample(20))
```

```python
import pandas as pd
import numpy as np
import unicodedata
import re

# ============================================================
# 1. CHARGEMENT DES DATASETS
# ============================================================

print("df_joueurs :", df_joueurs.shape)
print("df_final   :", df_final.shape)


# ============================================================
# 2. NORMALISATION DES NOMS (clé de merge)
# ============================================================

def normalize_name(name):
    """Lowercase, supprime accents, retire caractères spéciaux."""
    if pd.isna(name):
        return np.nan
    name = str(name).lower().strip()
    name = unicodedata.normalize("NFD", name)
    name = "".join(c for c in name if unicodedata.category(c) != "Mn")
    name = re.sub(r"[^a-z\s]", "", name)
    name = re.sub(r"\s+", " ", name).strip()
    return name

df_joueurs["player_clean"] = df_joueurs["player_name"].apply(normalize_name)
df_final["player_clean"]   = df_final["player_name"].apply(normalize_name)


# ============================================================
# 3. VÉRIFICATION DU TAUX DE MATCHING
# ============================================================

joueurs_set = set(df_joueurs["player_clean"].dropna())
final_set   = set(df_final["player_clean"].dropna())

communs     = joueurs_set & final_set
sans_stats  = joueurs_set - final_set
sans_joueur = final_set - joueurs_set

print(f"\nJoueurs en commun         : {len(communs)}")
print(f"Joueurs sans stats xG     : {len(sans_stats)}")
print(f"Stats xG sans fiche joueur: {len(sans_joueur)}")


# ============================================================
# 4. NETTOYAGE PRÉ-MERGE — résoudre les conflits de colonnes
# ============================================================

# 'position' existe dans les deux → renommer celle de df_final
df_final = df_final.rename(columns={
    "position"    : "position_stats",
    "player_name" : "player_name_stats"   # éviter le doublon _x/_y
})


# ============================================================
# 5. MERGE
# ============================================================

df_merged = pd.merge(
    df_joueurs,
    df_final,
    on="player_clean",
    how="left"    # ← garde tous les joueurs ; change en "inner" pour le modèle
)

print(f"\nDataset mergé : {df_merged.shape}")


# ============================================================
# 6. NETTOYAGE POST-MERGE
# ============================================================

# Supprimer player_name_stats (doublon) et l'id de df_final
df_merged = df_merged.drop(columns=["player_name_stats", "id"], errors="ignore")

# Réordonner les colonnes
cols_info  = ["player_id", "player_name", "player_clean", "team", "team_crest",
              "position", "dob", "nationality", "photo_url"]
cols_stats = ["games", "time", "goals", "xG", "assists", "xA", "shots",
              "key_passes", "yellow_cards", "red_cards", "position_stats",
              "team_title", "npg", "npxG", "xGChain", "xGBuildup", "league"]

cols_info  = [c for c in cols_info  if c in df_merged.columns]
cols_stats = [c for c in cols_stats if c in df_merged.columns]

df_merged = df_merged[cols_info + cols_stats]


# ============================================================
# 7. DATASET POUR LE MODÈLE DE PRÉDICTION
# ============================================================

# Uniquement les joueurs avec des stats xG complètes
df_model = df_merged.dropna(subset=["xG", "xA"]).copy()

print(f"\nDataset modèle : {df_model.shape}")

# Top 10 buteurs potentiels (xG)
print("\n--- Top 10 buteurs potentiels (xG) ---")
print(
    df_model[["player_name", "team", "xG", "npxG", "shots", "goals"]]
    .sort_values("xG", ascending=False)
    .head(10)
    .to_string(index=False)
)

# Top 10 passeurs potentiels (xA)
print("\n--- Top 10 passeurs potentiels (xA) ---")
print(
    df_model[["player_name", "team", "xA", "key_passes", "assists"]]
    .sort_values("xA", ascending=False)
    .head(10)
    .to_string(index=False)
)


# ============================================================
# 8. EXPORT
# ============================================================

df_merged.to_csv("df_merged_complet.csv", index=False)
df_model.to_csv("df_model_prediction.csv", index=False)

print("\n✅ Fichiers exportés :")
print("   → df_merged_complet.csv   (tous les joueurs)")
print("   → df_model_prediction.csv (joueurs avec stats xG, prêts pour le modèle)")
```
