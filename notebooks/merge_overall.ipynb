```python
import pandas as pd

# Charger le fichier en désactivant le mode low_memory
df = pd.read_csv('https://raw.githubusercontent.com/ismailoksuz/EAFC26-DataHub/refs/heads/main/data/players.csv', low_memory=False)
```

```python
df.head()
```

```python
df_overall = df[['short_name', 'overall']]

df_overall.head()
```

```python
olise = df_overall['short_name'] == "M. Olise"
df_overall[olise]
```

```python
df_joueurs = pd.read_csv('C:\\Users\\user\\Projet3\\WorldCup2026\\notebooks\\df_joueurs_niveau_equipe.csv')
df_joueurs.shape
```

```python
df_joueurs.head()
```

```python
# 1. On crée des colonnes temporaires uniquement pour la fusion, sans toucher aux originaux
df_joueurs['key'] = df_joueurs['nom_joueur'].str.split().str[-1]
df_overall['key'] = df_overall['short_name'].str.split().str[-1]

# 2. On fait le merge sur ces clés
df_final = pd.merge(
    df_joueurs, 
    df_overall[['key', 'overall']], # On utilise la clé pour le merge
    on='key', 
    how='left'
)

# 3. On supprime la clé temporaire
df_final = df_final.drop(columns=['key'])
```

```python
# 1. On s'assure que 'overall' est bien numérique pour pouvoir trier correctement
df_final['overall'] = pd.to_numeric(df_final['overall'], errors='coerce')

# 2. On trie par 'overall' décroissant (le meilleur en haut)
df_final = df_final.sort_values(by='overall', ascending=False)

# 3. On supprime les doublons en se basant sur le nom complet 
# (on ne splitte plus, on utilise le nom original pour garder la précision)
df_final = df_final.drop_duplicates(subset=['nom_joueur'], keep='first')

# 4. On réinitialise l'index
df_final = df_final.reset_index(drop=True)
```

```python
df_final.shape
```

```python
df_final.to_csv("df_overall_joueurs.csv")
```

```python
# 1. Calculer la moyenne des notes par équipe nationale
moyennes_equipes = df_final.groupby('equipe_nationale')['overall'].mean()

# 2. Convertir le résultat en DataFrame pour qu'il soit plus facile à manipuler et à trier
df_moyennes = moyennes_equipes.reset_index()

# 3. Renommer la colonne pour plus de clarté
df_moyennes = df_moyennes.rename(columns={'overall': 'moyenne_equipe'})

# 4. Trier par ordre décroissant (de la meilleure équipe à la moins bonne)
df_moyennes = df_moyennes.sort_values(by='moyenne_equipe', ascending=False)

# Afficher les 10 meilleures équipes
display(df_moyennes.head(25))
```

```python
# Remplace partout dans le DataFrame
df_final = df_final.fillna(50)
```

```python
# 1. Calculer la moyenne par équipe et l'associer à chaque ligne correspondante
# La fonction 'transform' garantit que chaque joueur reçoit la moyenne de son équipe
df_final['moyenne_equipe'] = df_final.groupby('equipe_nationale')['overall'].transform('mean')

# 2. Vérifier le résultat
# On affiche quelques colonnes pour voir l'intégration
display(df_final[['nom_joueur', 'equipe_nationale', 'overall', 'moyenne_equipe']].head())
```
