```python
import pandas as pd

# Charger le fichier en désactivant le mode low_memory
df = pd.read_csv('https://raw.githubusercontent.com/ismailoksuz/EAFC26-DataHub/refs/heads/main/data/players.csv', low_memory=False)
```

```python
df.head()
```

```python
df_overall = df[['short_name', 'overall']]

df_overall.head()
```

```python
olise = df_overall['short_name'] == "M. Olise"
df_overall[olise]
```

```python
df_joueurs = pd.read_csv('C:\\Users\\user\\Projet3\\WorldCup2026\\notebooks\\df_joueurs_niveau_equipe.csv')
df_joueurs.shape
```

```python
df_joueurs.head()
```

```python
# 1. On crée des colonnes temporaires uniquement pour la fusion, sans toucher aux originaux
df_joueurs['key'] = df_joueurs['nom_joueur'].str.split().str[-1]
df_overall['key'] = df_overall['short_name'].str.split().str[-1]

# 2. On fait le merge sur ces clés
df_final = pd.merge(
    df_joueurs, 
    df_overall[['key', 'overall']], # On utilise la clé pour le merge
    on='key', 
    how='left'
)

# 3. On supprime la clé temporaire
df_final = df_final.drop(columns=['key'])
```

```python
df_final.shape
```

```python
# 1. Filtrer pour ne garder que les joueurs de la France
# Vérifie bien que le nom de la colonne et la valeur correspondent (ex: "France" ou "France")
equipe_france = df_final[df_final['nom_joueur'] == 'Nico Paz']

# 2. Afficher la liste des joueurs
print("Liste des joueurs de l'équipe de France :")
print(equipe_france[['nom_joueur', 'poste', 'overall']])

# 3. Optionnel : Afficher le nombre total de joueurs trouvés
display(f"\nNombre total de joueurs dans l'effectif : {len(equipe_france)}")

df_final.duplicated().sum()
```

```python
# 1. On s'assure que 'overall' est bien numérique pour pouvoir trier correctement
df_final['overall'] = pd.to_numeric(df_final['overall'], errors='coerce')

# 2. On trie par 'overall' décroissant (le meilleur en haut)
df_final = df_final.sort_values(by='overall', ascending=False)

# 3. On supprime les doublons en se basant sur le nom complet 
# (on ne splitte plus, on utilise le nom original pour garder la précision)
df_final = df_final.drop_duplicates(subset=['nom_joueur'], keep='first')

# 4. On réinitialise l'index
df_final = df_final.reset_index(drop=True)
```
